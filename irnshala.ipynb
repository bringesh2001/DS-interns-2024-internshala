iterating through all pages

```python
from urllib.parse import urljoin

base_url = 'https://internshala.com'

job_title = []
company =  []
location = []
duration = []
stipend =  []
skills = []
perks = []
links=[]

for j in range(1, 5):
    webpage = requests.get(f'https://internshala.com/internships/data-science-internship/page-{j}/').text
    soup = BeautifulSoup(webpage, 'html.parser')
    container = soup.find_all('div', class_='internship_meta')
    link_elements = soup.find_all('a', class_='view_detail_button_outline')
    hrefs = [urljoin(base_url, a_tag['href']) for a_tag in link_elements if 'href' in a_tag.attrs]

    for i, element in enumerate(container):
        try:
            job_title.append(element.find('h3', class_='heading_4_5 profile').text.strip())
        except:
            job_title.append(np.nan)
        try:
            company.append(element.find('p').text.strip())
        except:
            company.append(np.nan)
        try:
            location.append(element.find('a').text.strip())
        except:
            location.append(np.nan)
        try:
            stipend.append(element.find('span', class_='stipend').text.strip())
        except:
            stipend.append(np.nan)
        try:
            item_bodies = element.find_all('div', class_='item_body')
            duration_found = False
            for item in item_bodies:
                if 'Months' in item.text.strip():
                    duration.append(item.text.strip())
                    duration_found = True
                    break
            if not duration_found:
                duration.append(np.nan)
        except:
            duration.append(np.nan)

    for url in hrefs:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        skills_and_perks = soup.find_all('h3')
        
        job_skills = []
        job_perks = []
        
        for element in skills_and_perks:
            if element.text == 'Skill(s) required':
                skills_div = element.find_next_sibling('div')
                if skills_div:
                    try:
                        job_skills = [span.text.strip() for span in skills_div.find_all('span')]
                    except:
                        job_skills = [np.nan]
                        
            elif element.text == 'Perks':
                perks_div = element.find_next_sibling('div')
                if perks_div:
                    try:
                        job_perks = [span.text.strip() for span in perks_div.find_all('span')]
                    except:
                        job_perks = [np.nan]
        
        skills.append(job_skills)
        perks.append(job_perks)

# Ensuring the lengths of all lists match
min_length = min(len(job_title), len(company), len(location), len(duration), len(stipend), len(skills), len(perks))
job_title = job_title[:min_length]
company = company[:min_length]
location = location[:min_length]
duration = duration[:min_length]
stipend = stipend[:min_length]
skills = skills[:min_length]
perks = perks[:min_length]

# Print lengths for verification
print("Job Title:", len(job_title))
print("Company:", len(company))
print("Location:", len(location))
print("Duration:", len(duration))
print("Stipend:", len(stipend))
print("Skills:", len(skills))
print("Perks:", len(perks))
```

FINAL TABLE.

```python
df=pd.DataFrame({'job_title':job_title,
    'company':company,
    'location':location,
    'duration':duration,
    'stipend':stipend,
    'skills':skills,
    'perks':perks
    
    })
df
```
